```
#insertion sort
import random
A = [random.randrange(1, 101) for _ in range(101)] 
def insertionSort(A): 
    for i in range(1, len(A)): 
        key = A[i] 
        j = i-1
        while j >=0 and key < A[j] : 
                A[j+1] = A[j] 
                j -= 1
        A[j+1] = key 
insertionSort(A) 
print ("Sorted array is:") 
for i in range(len(A)): 
    print ("%d," %A[i],end=' ') 
```

```
#quick sort

import random
A = [random.randrange(1, 101) for _ in range(101)] 
def quickSort(A):
  quick(A,0,len(A)-1)
def quick(A,first,last):
  if first<last:
      split = partition(A,first,last)
      quick(A,first,split-1)
      quick(A,split+1,last)
def partition(A,first,last):
  pivot = A[first]
  left = first+1
  right = last
  done = False
  while not done:
      while left <= right and A[left] <= pivot:
          left = left + 1
      while A[right] >= pivot and right >= left:
          right = right -1
      if right < left:
          done = True
      else:
          temp = A[left]
          A[left] = A[right]
          A[right] = temp
  temp = A[first]
  A[first] = A[right]
  A[right] = temp
  return right
quickSort(A)
print(A)
```
